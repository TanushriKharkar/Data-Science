# Feature Scaling

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Online Sales Data.csv")
dataset.head(5)
```

## 1. Standardization

```python
dataset.isnull().sum()
```

```python
dataset["Unit Price"].fillna(dataset["Unit Price"].mean(), inplace=True)
dataset.head(5)
```

```python
sns.distplot(dataset["Unit Price"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss =StandardScaler()
ss.fit(dataset[["Unit Price"]])
```

```python
ss.transform(dataset[["Unit Price"]])
```

```python
dataset["Unit Price_ss"] = pd.DataFrame(ss.transform(dataset[["Unit Price"]]), columns=["x"])
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.displot(dataset["Unit Price"])
plt.title("Before")
plt.show()
```

```python
sns.displot(dataset["Unit Price_ss"])
plt.title("After")
plt.show()
```

```python
#Data got shrink but nature remain as it is
#due to scaling magnitude reduces
```

## 2. Normalisation

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["Unit Price"]])
```

```python
dataset["Unit Price_min"] = ms.transform(dataset[["Unit Price"]])
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["Unit Price"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Unit Price_min"])

plt.show()
```
