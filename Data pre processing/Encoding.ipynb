# One Hot Encoding

```python
#it convert categorical data into numerical data
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv("Online Sales Data.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset["Unit Price"].fillna(dataset["Unit Price"].mode()[0], inplace=True)    
#this will fill the data in unit price column
```

```python
en_data = dataset[["Region", "Payment Method"]]
en_data
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ar = ohe.fit_transform(en_data)
#it firstly see, read and then it transform the data into numerical form
```

```python
ohe = OneHotEncoder(drop="first")
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar,columns=["Region_Asia", "Region_Europe", "Region_North America",
                        "Payment Method_Credit Card","Payment Method_Debit Card",
                        "Payment Method_PayPal"])
```
