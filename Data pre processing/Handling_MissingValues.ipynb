# Handling Missing values by dropping rows or columns

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Online Sales Data.csv")           #as it already in the same folder
dataset.head(4)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
sns.heatmap(dataset.isnull())
plt.show()
```

```python
#if there is large number of data missing in a column drop that column
#if there is large number of data missing in a row drop that row
dataset.drop(columns=["Product Category"])
```

```python
dataset.dropna(inplace=True)          #this removes the missing values rows
                                      #inplace=True make the updation we made
```

```python
#all missing values get deleted
dataset.isnull().sum()
```

```python
sns.heatmap(dataset.isnull())
plt.show()
```

# Handling Missing values by imputing categorical data

```python
dataset = pd.read_csv("Online Sales Data.csv")           #as it already in the same folder
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset.fillna(10)                       #this fill 10 in all the Nan data
```

```python
dataset.info()
```

```python
dataset.fillna(method="bfill", axis=1)               #this will fil backward data
```

```python
dataset["Product Category"].mode()
```

```python
dataset["Product Category"].fillna(dataset["Product Category"].mode()[0], inplace=True)
dataset
```

```python
for i in dataset.select_dtypes(include="object").columns:
    dataset[i].fillna(dataset[i].mode()[0], inplace=True)
```

# Handling Missing values by Scikit learn

```python
dataset = pd.read_csv("Online Sales Data.csv")           #as it already in the same folder
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include="float64").columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy="mean")
ar = si.fit_transform(dataset[['Transaction ID', 'Unit Price', 'Total Revenue']])

new_dataset = pd.DataFrame(ar, columns=dataset.select_dtypes(include="float64").columns)
#this fill the missing value data of given columns
```

```python
new_dataset.isnull().sum()
```

```python
new_dataset
```
