# Pie Chart/Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Basic Structure
x = [10,20,30,40]
y = ["python","C", "C++","Java"]

plt.pie(x,labels= y)
plt.show()
```

```python
x = [10,20,30,40]
y = ["python","C", "C++","Java"]
ex = [0.4, 0.0, 0.0, 0.0 ]
c = ['r','b', 'y', 'm']

plt.pie(x,labels= y, explode=ex, colors=c, autopct="%0.1f%%")
plt.show()
```

```python
x = [10,20,30,40]
y = ["python","C", "C++","Java"]
ex = [0.4, 0.0, 0.0, 0.0 ]
c = ['r','b', 'y', 'm']

plt.pie(x,labels= y, explode=ex, colors=c, autopct="%0.1f%%", shadow=True, radius=0.6,labeldistance= 1.4 )
plt.show()
```

```python
x = [10,20,30,40]
y = ["python","C", "C++","Java"]
ex = [0.4, 0.0, 0.0, 0.0 ]
c = ['r','b', 'y', 'm']

plt.pie(x,labels= y, explode=ex, colors=c, autopct="%0.1f%%", startangle=0,
        textprops={"fontsize":15}, counterclock=False,shadow=True, wedgeprops={"linewidth":4,})

plt.title("Pie Chart")
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

## Multiple Pie Chart

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["python","C", "C++","Java"]
c = ['r','b', 'y', 'm']

plt.pie(x,labels= y, radius=1.5)
plt.pie(x1, labels= y, radius=1, colors=c)

plt.show()
```

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["python","C", "C++","Java"]

plt.pie(x,labels= y, radius=1.5)
plt.pie([1], colors='w')

plt.show()
```

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["python","C", "C++","Java"]
c = ['r','b', 'y', 'm']

plt.pie(x,labels= y, radius=1.5)
cr = plt.Circle(xy=(0,0), radius=1, facecolor='w')
plt.gca().add_artist(cr)

plt.show()
```
