# Step Plot

```python
import matplotlib.pyplot as plt
```

```python
# Basic structure
#it is used in communication signal
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.step(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x,y, color='y', marker="o", ms=10, mfc='b', label="Python" )

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc=2)
plt.grid()
plt.show()
```
