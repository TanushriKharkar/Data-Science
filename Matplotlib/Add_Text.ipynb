# Text adding methods

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("title", fontsize=15)
plt.xlabel("x-axis", fontsize=15)
plt.ylabel("y-axis", fontsize=15)

plt.text(2,5, "python", fontsize=15, style="italic", bbox={"facecolor": "yellow"})

plt.annotate("python", xy=(2,1), xytext=(4,4), arrowprops=dict(facecolor="black", shrink=100))

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x,y)

plt.title("title", fontsize=15)
plt.xlabel("x-axis", fontsize=15)
plt.ylabel("y-axis", fontsize=15)

plt.legend(["up"], facecolor="yellow", edgecolor="magenta", framealpha=0.5, shadow=True)
plt.show()
```
