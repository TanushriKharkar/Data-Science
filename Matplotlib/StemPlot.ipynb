# Stem Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Basic Structure
x = [1,2,3,4,5,6]
y = [2,3,2,4,8,5]

plt.stem(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,3,2,4,8,5]

plt.stem(x,y, linefmt=":", markerfmt= "r*", bottom=0, basefmt='g', label="chocolate")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,3,2,4,8,5]

plt.stem(x,y, orientation="horizontal")
plt.show()
```
