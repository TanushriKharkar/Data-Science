# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Basic structure
x = ["Python", "C", "C++", "Java"]
y = [89, 78, 80, 93]

plt.bar(x,y)
plt.show()
```

```python
x = ["Python", "C", "C++", "Java"]
y = [89, 78, 80, 93]

plt.xlabel("Languages", fontsize = 15 )
plt.ylabel("Percent", fontsize = 15 )
plt.title("Bar plot", fontsize = 15)

c = ['y', 'b', 'm', 'g']

plt.bar(x,y, width = 0.4, color = c)
plt.show()
```

```python
x = ["Python", "C", "C++", "Java"]
y = [89, 78, 80, 93]

plt.xlabel("Languages", fontsize = 15 )
plt.ylabel("Percent", fontsize = 15 )
plt.title("Bar plot", fontsize = 15)

c = ['y', 'b', 'm', 'g']

plt.bar(x,y, width = 0.4, color = c, edgecolor = 'r', linewidth = 3, linestyle =":", alpha = 0.5  )
plt.show()
```

```python
x = ["Python", "C", "C++", "Java"]
y = [89, 78, 80, 93]
z = [20,40,60,80]

plt.xlabel("Languages", fontsize = 15 )
plt.ylabel("Percent", fontsize = 15 )
plt.title("Bar plot", fontsize = 15)

plt.bar(x,y, width = 0.4, color = 'r', label = "Bar1")
plt.bar(x,z, width = 0.4, color = 'y', label = "Bar2")
plt.legend()
plt.show()
```

```python
x = ["Python", "C", "C++", "Java"]
y = [89, 78, 80, 93]
z = [20,40,60,80]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Languages", fontsize = 15 )
plt.ylabel("Percent", fontsize = 15 )
plt.title("Bar plot", fontsize = 15)

plt.bar(p,y, width, color = 'r', label = "Bar1")
plt.bar(p1,z, width, color = 'y', label = "Bar2")

plt.xticks(p+width/2, x, rotation = 5)
plt.legend()
plt.show()
```

```python
#horizontal bar graph
x = ["Python", "C", "C++", "Java"]
y = [89, 78, 80, 93]
z = [20,40,60,80]

plt.xlabel("Percent", fontsize = 15 )
plt.ylabel("Languages", fontsize = 15 )
plt.title("Bar plot", fontsize = 15)

plt.barh(x,y, color = 'r', label = "Bar1")
plt.barh(x,z,color = 'y', label = "Bar2")
plt.legend()
plt.show()
```
