# Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
# Basic structure
x = [1,2,3,4,5]
area = [1,2,3,4,5]

plt.stackplot(x,area)
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,1,4,5,6]
area3 = [1,3,2,4,2]

l = ["Area1", "Area2", "Area3"]

plt.stackplot(x,area1,area2, area3, labels=l, colors=['r','g','y'])

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,1,4,5,6]
area3 = [1,3,2,4,2]

l = ["Area1", "Area2", "Area3"]

plt.stackplot(x,area1,area2, area3, labels=l, colors=['r','g','y'], baseline="sym")

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.grid()
plt.legend()
plt.show()
```
