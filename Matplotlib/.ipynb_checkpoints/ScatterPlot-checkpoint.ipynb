# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# basic structure
x = [1,2,3,4,5,6,7]
y = [2,4,6,3,8,9,1]

plt.scatter(x,y)
plt.show()
```

```python
Day = [1,2,3,4,5,6,7]
No = [2,4,6,3,8,9,1]

plt.scatter(Day, No, color = 'y')
plt.xlabel("Day", fontsize = 15 )
plt.ylabel("No.", fontsize = 15 )
plt.title("Scatter plot", fontsize = 15)

plt.show()
```

```python
Day = [1,2,3,4,5,6,7]
No = [2,4,6,3,8,9,1]
colors = ['r', 'y', 'g', 'b', 'm', 'y', 'm']

sizes = [100,290,400,150,100,170,460]
plt.scatter(Day, No, c=colors, s=sizes, marker="*", edgecolor = 'b', linewidth= 2)
plt.xlabel("Day", fontsize = 15 )
plt.ylabel("No.", fontsize = 15 )
plt.title("Scatter plot", fontsize = 15)

plt.show()
```

```python
Day = [1,2,3,4,5,6,7]
No = [2,4,6,3,8,9,1]
colors = [10,49,20,60,30,76,84]

sizes = [100,290,400,150,100,170,460]
plt.scatter(Day, No, c=colors, s=sizes, cmap="viridis")
t=plt.colorbar()
t.set_label("ColorBar", fontsize=15)
plt.xlabel("Day", fontsize = 15 )
plt.ylabel("No.", fontsize = 15 )
plt.title("Scatter plot", fontsize = 15)

plt.show()
```

```python
Day = [1,2,3,4,5,6,7]
No = [2,4,6,3,8,9,1]
No2 = [4,6,1,8,3,9,3]
colors = [10,49,20,60,30,76,84]

sizes = [100,290,400,150,100,170,460]
plt.scatter(Day, No, c=colors, s=sizes, cmap="viridis")
plt.scatter(Day, No2, c='y', s=sizes, cmap="viridis")

t=plt.colorbar()
t.set_label("ColorBar", fontsize=15)
plt.xlabel("Day", fontsize = 15 )
plt.ylabel("No.", fontsize = 15 )
plt.title("Scatter plot", fontsize = 15)

plt.show()
```
