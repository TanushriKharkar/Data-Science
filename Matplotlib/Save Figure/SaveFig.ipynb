# Save Figure

```python
import matplotlib.pyplot as plt
```

```python
# basic Structure
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y)
plt.savefig("line")

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y)
plt.savefig("line1", dpi = 2000)     #for good quality

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y)
plt.savefig("line2",dpi = 2000, facecolor='r')

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y)
plt.savefig("line3.pdf", dpi = 2000)     

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y)
plt.savefig("line5.jpg", dpi = 2000)     

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y)
plt.savefig("line6", dpi = 2000,facecolor='r', transparent=True)     

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x,y)
plt.savefig("line7", dpi = 2000,facecolor='r',bbox_inches="tight" )     

plt.show()
```
