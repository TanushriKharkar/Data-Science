# Box Plot

```python
import matplotlib.pyplot as plt
```

```python
# Basic Structure
x = [10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, notch=True)

plt.show()
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x, vert=False, patch_artist=True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x, widths=0.5)
plt.show()
```

```python
x = [10,20,30,40,50,60,70, 120]
plt.boxplot(x, labels=["python"], showmeans=True, sym='g+',
            boxprops=dict(color='r'), capprops=dict(color='b'),
           flierprops=dict(markeredgecolor='m'))
plt.show()
```

```python
x = [10,20,30,40,50,60,70, 120]
plt.boxplot(x, labels=["python"], showmeans=True, whis=2)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,]
x1 = [10,20,40,50,30,60,90]

y=[x,x1]
plt.boxplot(y, labels=["python", "Java"], showmeans=True, sym='g+',
            boxprops=dict(color='r'), capprops=dict(color='b'))

plt.show()
```
