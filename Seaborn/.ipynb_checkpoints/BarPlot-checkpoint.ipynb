# Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
y1 = sns.load_dataset("penguins")
y1
```

```python
sns.set(style="dark")
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=y1, hue="sex", order= order_1,
           hue_order=["Female", "Male"], palette="icefire", saturation=1, 
           errcolor='y',capsize=0.5)
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm", data=y1, color='y')
plt.show()
```

```python
sns.barplot(x="bill_depth_mm", y="bill_length_mm", data=y1, orient='h')
plt.show()
```
