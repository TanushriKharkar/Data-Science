# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
 y1 = sns.load_dataset("penguins").head(20)
y1
```

```python
m={"Male":"*", "Female":"o"}
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=y1, hue="sex", style="sex",
               sizes=(90,60), palette="gist_ncar", markers=m)
                
plt.show()
```
