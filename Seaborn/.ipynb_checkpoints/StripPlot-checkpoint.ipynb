# Strip Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var= sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day", y="total_bill", data=var,hue="sex", palette="hsv_r",
             linewidth=1,size=5, alpha=0.5)
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=var,hue="sex", palette="hsv_r",
             linewidth=1, jitter=2)
plt.show()
```

```python
sns.stripplot( y=var["total_bill"])
plt.show()
```
