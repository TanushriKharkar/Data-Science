# Cat Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var= sns.load_dataset("tips")
var
```

```python
sns.catplot(x="tip", y="size", data=var, hue="sex", palette="brg_r", height=5)
plt.show()
```

```python
sns.catplot(x="tip", y="size", data=var, hue="sex", palette="brg_r", height=5,
           kind="point")
plt.show()
```
