# Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var= sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="time")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, linewidth=3, palette="twilight_shifted")
plt.show()
```

```python
sns.violinplot(x="time", y="total_bill", data=var,order=["Dinner", "Lunch"], inner="stick")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=var, hue="sex", split=True, scale="width")
plt.show()
```

```python
sns.violinplot(y="day", x="total_bill", data=var, hue="time")
plt.show()
```

```python
sns.violinplot(y=var["total_bill"])
plt.show()
```
