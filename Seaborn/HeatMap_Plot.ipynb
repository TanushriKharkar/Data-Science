# HeatMap Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var, vmin=0, vmax=20, cmap="gist_heat", annot=True)
plt.show()
```

```python
data = sns.load_dataset("anagrams")
x= data.drop(columns=["attnr"], axis=1).head(10)
x
```

```python
sns.heatmap(x)
plt.show()
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
y={"fontsize":15, "color":'y'}
v =sns.heatmap(var1, vmin=0, vmax=20, cmap="gist_heat", annot=True, annot_kws=y,
            linewidth=5, linecolor='b', cbar=False, yticklabels=False, 
            xticklabels=False)
v.set(xlabel="python", ylabel="yupp")
sns.set(font_scale=5)
plt.show()
```

```python
ar = np.array([["a1","a2","a3","a4","a5"],
              ["b1","b2","b3","b4","b5"]])
ar
```

```python
sns.heatmap(var1, vmin=0, vmax=20, cmap="gist_heat", annot=ar, fmt="s")
plt.show()
```
