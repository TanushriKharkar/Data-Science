# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# basic structure of matplotlib
var_1 = [1,2,3,4,5,6,7]
var_2 = [2,3,4,1,5,6,7]

plt.plot(var_1, var_2)
plt.show()
```

```python
# basic structure of seaborn
var_1 = [1,2,3,4,5,6,7]
var_2 = [2,3,4,1,5,6,7]

x = pd.DataFrame({"var_1":var_1, "var_2":var_2})

sns.lineplot(x="var_1", y="var_2", data=x)

plt.show()
```

```python
var_1 = [1,2,3,4,5,6,7]
var_2 = [2,3,4,1,5,6,7]

sns.lineplot(x=var_1, y=var_2)

plt.show()
```
